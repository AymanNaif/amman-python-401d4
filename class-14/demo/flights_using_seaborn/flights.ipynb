# Visualization with Seaborn Class

### USe matlptlib to plot simple graph

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([3,7,4,18,-1])
```

```python
plt.plot([2,6,10,16,26])
```

```python
def plot_lines(y1, y2):
    plt.figure() # Holder of the plot
    plt.plot(y1, label='y1 data') # in blue
    plt.plot(y2, label='y2 data') # in orange
    plt.title('Demo data for Visualization')
    plt.xlabel('Fake indeces')
    plt.ylabel('Arrays data')
    plt.legend() # Show the labels
    plt.show() # Show the plot

plot_lines([3,7,4,18,-1], [2,6,10,16,26,40, 70])
```

## Seaborn

```python
import seaborn as sns
```

```python
sns.set()
```

```python
plot_lines([3,7,4,18,-1], [2,6,10,16,26,40, 70])
```

## Seaborn data visualized

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
sns.relplot(x='year', y='passengers', data=flights)
```

```python
sns.relplot(x='month', y='passengers', data=flights)
```

```python
data_grouped_by_year = flights.groupby('year', as_index=False).mean()
```

```python
data_grouped_by_year
```

```python
data_grouped_by_month = flights.groupby('month', as_index=False).mean()
```

```python
data_grouped_by_month
```

```python
sns.relplot(x='month', y='passengers', data=data_grouped_by_month)
```

```python
sns.relplot(x='month', y='passengers', kind='line', data=data_grouped_by_month)
```

### BarPlot

```python
sns.barplot(x='month', y='passengers', data=data_grouped_by_month)
```

```python
plt.pie(data_grouped_by_month['passengers'])
```
